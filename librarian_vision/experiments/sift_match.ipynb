```python
file_path = "./match_data/"
query_img = file_path + "cc.jpg"
target_img = file_path + "id_22.png"


import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img1 = cv.imread(query_img,cv.IMREAD_GRAYSCALE)          # queryImage
img2 = cv.imread(target_img,cv.IMREAD_GRAYSCALE) # trainImage
# Initiate SIFT detector
sift = cv.SIFT_create()
# find the keypoints and descriptors with SIFT
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)
# FLANN parameters
FLANN_INDEX_KDTREE = 1
index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
search_params = dict(checks=50)   # or pass empty dictionary
flann = cv.FlannBasedMatcher(index_params,search_params)
matches = flann.knnMatch(des1,des2,k=2)
# print(matches)
# Need to draw only good matches, so create a mask
matchesMask = [[0,0] for i in range(len(matches))]
# ratio test as per Lowe's paper
match_score = 0
for i,(m,n) in enumerate(matches):
    if m.distance < 0.45*n.distance:
        matchesMask[i]=[1,0]
        match_score = match_score + 1
# print(matchesMask)
print(f"match_score:{match_score}")
draw_params = dict(matchColor = (0,255,0),
                   singlePointColor = (255,0,0),
                   matchesMask = matchesMask,
                   flags = cv.DrawMatchesFlags_DEFAULT)
img3 = cv.drawMatchesKnn(img1,kp1,img2,kp2,matches,None,**draw_params)
plt.imshow(img3,),plt.show()
```
