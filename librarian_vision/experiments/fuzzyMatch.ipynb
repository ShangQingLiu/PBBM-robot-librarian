```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
print(fuzz.ratio("Catherine M Gitau","Catherine Gitau"))
#91
print(fuzz.partial_ratio("Catherine M. Gitau","Catherine Gitau"))
#100
```

```python
print(fuzz.ratio("Catherine M Gitau","Gitau Catherine"))
#55
print(fuzz.partial_ratio("Catherine M. Gitau","Gitau Catherine"))
#60
```

```python
fuzz.token_sort_ratio("Catherine Gitau M.", "Gitau Catherine")
#94
```

```python
import pandas as pd

df = pd.read_csv('../librarian_common/librarian_resources/books.csv')
title_list = df["title"].tolist()

test_phrase = "gestützte CAE"
best_score = 0
best_index = -1

for index, title in enumerate(title_list):
    score = fuzz.partial_ratio(title, test_phrase)
    print(f"score:{score},title: {title}, test_phrase:{test_phrase}") 
    if best_score < score:
        best_score = score
        best_index = index

print(f"best index:{best_index}")
print(f"best score:{best_score}")
print(title_list[best_index])
```
