```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

book_blue = cv2.imread("book_blue.png")
book_yellow = cv2.imread("book_yellow.png")

plt.imshow(cv2.cvtColor(book_blue, cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(book_yellow, cv2.COLOR_BGR2RGB))
```

```python
book_blue_hsv = cv2.cvtColor(book_blue, cv2.COLOR_BGR2HSV)
hist_blue, _ = np.histogram(book_blue_hsv[0], bins=8)
plt.plot(hist_blue, label="blue book")
plt.legend()
```

```python
book_yellow_hsv = cv2.cvtColor(book_yellow, cv2.COLOR_BGR2HSV)
hist_yellow, _ = np.histogram(book_yellow_hsv[0], bins=8)
plt.plot(hist_yellow, label="yellow book")
plt.legend()
```

# Show the similarity

the histograms have a low similarity because pvalue=0.02

```python
from scipy.stats import ks_2samp

ks_2samp(hist_blue, hist_yellow)
```

```python
import glob
import cv2
import numpy as np
import re

all_hists = []
spines = glob.glob("spines/*")
spines = list(map(lambda s: [int(re.findall("\d+", s)[0]), s], spines))
spines.sort()

for i, file in spines:
    print(file)
    img = cv2.imread(file)
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hist, _ = np.histogram(hsv[0], bins=8)
    hist = hist / np.linalg.norm(hist)
    all_hists.append(hist)

np.savetxt("hue.txt", all_hists)
    
```
